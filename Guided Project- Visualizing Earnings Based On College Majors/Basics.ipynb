# Reading the dataset

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

```python
raw_data_count = recent_grads.index
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.index
print(raw_data_count)
recent_grads.describe()
```

```python
plot1 = recent_grads.plot(x = 'Sample_size' , y = 'Median', kind = 'scatter',title='Sample size vs Median', figsize=(6,6))

plot2 = recent_grads.plot(x = 'Sample_size' , y = 'Unemployment_rate', kind = 'scatter',title='Sample size vs Unemployment_rate', figsize=(6,6))

plot3 = recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter',title='Full_time vs Median', figsize=(6,6))

plot4 = recent_grads.plot(x = 'ShareWomen' , y = 'Unemployment_rate', kind = 'scatter',title='ShareWpmen vs Unemployment_rate', figsize=(6,6))

plot5 = recent_grads.plot(x = 'Men' , y = 'Median', kind = 'scatter',title='Mean vs Median', figsize=(6,6))

plot6 = recent_grads.plot(x = 'Women' , y = 'Median', kind = 'scatter',title='Women vs Medina', figsize=(6,6))

plt.show()


```

```python
cols = [ 'Sample_size','Median','Employed' ,'Full_time' ,'ShareWomen' , 'Unemployment_rate', 'Men', 'Women' ]
fig = plt.figure(figsize = (5,15))
for val in range(1,9):
    ax = fig.add_subplot(8,1,val)
    ax = recent_grads[cols[val-1]].plot(kind="hist",bins = 20,title=cols[val-1], rot = 30)
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.show()
```

```python
from pandas.tools.plotting import scatter_matrix
cols1 = ['Sample_size','Median']
cols2 = ['Sample_size','Median','Unemployment_rate']
ax = scatter_matrix(recent_grads[cols1], figsize=(10,10))
ax = scatter_matrix(recent_grads[cols2], figsize=(10,10))
plt.show()
```
